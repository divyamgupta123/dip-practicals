```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Reading Image
img = cv2.imread('/content/ml.jpg')
```

```python
print("Image Shape - ",img.shape)
```

```python
#BGR Format 
print("Image in BGR Format - ")
plt.imshow(img)
plt.show()
```

```python
#Converting to RBG Format

imgRGB = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

print("Image in RGB Format - ")
plt.imshow(imgRGB)
plt.show()
```

```python
#Converting to HSV Format

imgHSV = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

print("Image in HSV Format - ")
plt.imshow(imgHSV)
plt.show()
```

```python
#Converting to HLS Format

imgHLS = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)

print("Image in HLS Format - ")
plt.imshow(imgHLS)
plt.show()
```

```python
#Converting to Gray Format

imgGray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

print("Image in Gray Format - ")
plt.imshow(imgGray)
plt.show()
```

```python
# Blue Channel only
imageCopy = imgRGB.copy()

imageCopy[:,:,1] = 0
imageCopy[:,:,0] = 0

print("Only Blue Channel - ")
plt.imshow(imageCopy)
plt.show()
```

```python
# Red Channel only
imageCopy = imgRGB.copy()

imageCopy[:,:,1] = 0
imageCopy[:,:,2] = 0

print("Only Red Channel - ")
plt.imshow(imageCopy)
plt.show()
```

```python
# Green Channel only
imageCopy = imgRGB.copy()

imageCopy[:,:,0] = 0
imageCopy[:,:,2] = 0

print("Only Green Channel - ")
plt.imshow(imageCopy)
plt.show()
```

```python
# Addition of 2 Channels - Blue and Red

# Blue Channel only
imageBlue = imgRGB.copy()

imageBlue[:,:,0] = 0
imageBlue[:,:,1] = 0

# Red Channel only
imageRed = imgRGB.copy()

imageRed[:,:,1] = 0
imageRed[:,:,2] = 0

print("Addition Blue and Red Channel - ")
plt.imshow(imageRed+imageBlue)
plt.show()
```
